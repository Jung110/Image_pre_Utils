```python
dir_path = r"/content/drive/MyDrive/final_project/sample/*.mp4"
filelist = glob.glob(dir_path)
filelist
```

```python
for i in filelist:
    vidcap = cv2.VideoCapture(i)
    print('Frame height:', int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT)))
```

```python
# 영상의 의미지를 연속적으로 캡쳐할 수 있게 하는 class
# 영상이 있는 경로

for i in filelist:
    vidcap = cv2.VideoCapture(i)
    name = i.split("/")[-1].split(".")[0]
    count = 0
    n = 1000 # 프레임지정

    while(vidcap.isOpened()):
        ret, image = vidcap.read()
        # 이미지 사이즈 960x540으로 변경
        # image = cv2.resize(image, (960, 540))
        
        if(int(vidcap.get(1)) % n == 0): # n프레임당 하나씩 이미지 추출
            print('Saved frame number : ' + str(int(vidcap.get(1))))
            frame = int(vidcap.get(1))
            cv2.imwrite(f"/content/drive/MyDrive/final_project/test/{name}_frame_{frame}.png", image) # 저장경로
            #print('Saved frame%d.jpg' % count)
            count += 1
            if frame == (int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT)) // n)*n:
                break
    vidcap.release()
```
