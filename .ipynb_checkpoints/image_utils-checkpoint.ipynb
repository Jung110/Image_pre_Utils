```python
# change
import os
import fnmatch
import glob
import json
import pandas as pd
from typing import Sequence
from typing import Dict
'''
ffmpeg 설치 필요
'''
import cv2
from google.colab.patches import cv2_imshow
```

```python
def find_specific_format_file(file_dir , file_format):
    '''
    file_dir 경로에 있는 파일중 
    file_format 형식을 가지는 파일의 경로 리스트를 반환
    '''
    try:
        file_names =os.listdir(file_dir)
        results = []
        for file_name in file_names:
            if fnmatch.fnmatch(file_name,"*."+file_format):
                results.append(file_dir+'/'+file_name)

        return results
    except(Exception):
        print('Error in find_specific_format_file function')
        return None
            
```

```python
def replace_class(file_path, save_dir_name, old_txt , new_txt , sep = ' ',check = False):
    '''
    file_path 에 있는 데이터 클래스를 저장하는 파일을 sep를 기준으로 
    읽어 old_txt 인 클래스를 new_txt값으로 바꾸어 준다.
    그 후 해당 파일을 save_dir_name에 저장한다.
    check 여부에 따라 해당 과정을 출력할 수 있다.
    '''
    save_path = make_save_path(file_path , save_dir_name)
    try:
        with open(file_path,'r') as f:
            with open(save_path,'w') as w:
                if check:
                    print('file_path: ',file_path,end = '\n')
                for line in f:
                    line_split_list = line.split(sep)
                    class_label = line_split_list[0]
                    class_label = class_label.replace(old_txt,new_txt)
                    line_split_list = [class_label] + line_split_list[1:]
                    line = sep.join(line_split_list)
                    if check:
                        print(f'line{count}: {line}')
                        count +=1
                    w.writelines(line)
                    
    except(Exception):
        print('Error in replace_class function')
        return None
        
            
        
```

```python
def make_save_path(file_path,save_dir_name):
    '''
    file_path 경로에 save_dir_name의 경로가 없는 경우  
    save_dir_name 생성하는 함수
    '''
    file_path_split_list = file_path.split('/')
    file_name = file_path_split_list[-1]
    save_file_path = '/'.join(file_path_split_list[:-1]+[save_dir_name])
    if not os.path.exists(save_file_path):
        os.makedirs(save_file_path)
        
    save_file_path = save_file_path+'/'+file_name
    return save_file_path
```

```python
def replace_class_in_dir(file_dir ,file_format,save_dir_name,old_txt , new_txt , sep = ' ' , check = False):
    '''
    file_path 에 있는 데이터 클래스를 저장하는 파일을 sep를 기준으로 
    읽어 old_txt 인 클래스를 new_txt값으로 바꾸어 준다.
    그 후 해당 파일을 save_dir_name에 저장한다.
    check 여부에 따라 해당 과정을 출력할 수 있다.
    '''
    file_path_list = find_specific_format_file(file_dir ,file_format)
    count = 1 
    for file_path in file_path_list:
        if check:
            print(f'{count}번째')
            count+=1
        replace_class(file_path,save_dir_name,old_txt,new_txt,sep,check)
```

```python
def compare_each_format_count(format1 : str,
                             format2 : str,
                             path : str) -> None:
    """
    각각 format1, format2 형식인 이름은 같고 format만 다른 파일의 이름을 추출
    위를 이용하여 어노테이션이 없는 이미지 데이터나 어노테이션만 있는 이미지 데이터를 알아낼 수 있다. 
    """
    
    format1_list = set()
    format2_list = set()
    
    for (file_dir ,_, file_names) in os.walk(path):
        for name  in file_names:
            name_list = name.split('.')
            if name_list[-1] == format1:
                format1_list.add(name_list[0])
            else:
                format2_list.add(name_list[0])
    if(len(format1_list) >= len(format2_list)):
        print(format1_list - format2_list ,format1 , 'is more')
    else:
        print(format2_list - format1_list,format2, 'is more')
```
